```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

# Fetch Nvidia data from Yahoo Finance for 10 years
df = yf.download("NVDA", start="2016-03-15", end="2026-03-15")
df.head()

# Calculate daily returns and annualized volatility
df['Returns'] = df['Close'].pct_change()
df['Volatility'] = df['Returns'].rolling(window=252).std() * np.sqrt(252)

# Plot the Nvidia prices
plt.figure(figsize=(14, 7))
plt.subplot(3, 1, 1)
plt.plot(df.index, df['Close'], label='Nvidia Prices', color='blue')
plt.title('Nvidia Prices')
plt.ylabel('Price')
plt.legend()

# Plot Nvidia daily stock returns
plt.subplot(3, 1, 2)
plt.plot(df.index, df['Returns'], label='Nvidia Returns', color="green")
plt.title('Nvidia Returns')
plt.ylabel('Return')
plt.legend()

# Plot the annualized volatility
plt.subplot(3, 1, 3)
plt.plot(df.index, df['Volatility'], label='Annualized Volatility', color='red')
plt.title('Nvidia Annualized Volatility')
plt.ylabel('Annualized Volatility')
plt.xlabel('Year')
plt.legend()

plt.tight_layout()
plt.show()

```
